```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
drinks = pd.read_csv("http://bit.ly/drinksbycountry")
```

```python
type(drinks)
```

```python
drinks['continent']
```

```python
drinks2 = drinks.groupby(['continent']).size()
```

```python
drinks2.plot.bar()
plt.show
```

```python
drinks.groupby(['continent']).mean().plot(kind="bar", figsize=(10,8), fontsize=10, rot=45)
```

```python
drinks.boxplot(column="beer_servings", by = "continent", figsize=(10,8), fontsize=10)
```
